```python
# importing the necessary libraries

import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

```python
class Image_Manipulation:
    
    def load_image(self):
        # function to load the image
        path = input('Enter the image path')
        self.img = io.imread(path)


    def horizontal_flip(self):
        # function for horizontal flipping
        print('Original image is')
        self.show()
        self.img = np.fliplr(self.img)
        print("Flipped image is")
        self.show()
    
    def vertical_flip(self):
        # function for vertical flipping
        print('Original image is')
        self.show()
        self.img = np.flipud(self.img)
        print("Flipped image is")
        self.show()
    
    def invert_color(self):
        # function for inverting image
        print('Original image is')
        self.show()
        mask = np.full(self.img.shape, 255)
        self.img = mask - self.img
        self.img.astype(np.uint8)
        print("Inverted image is")
        self.show()

    
    def to_grayscale(self):
        # function to convert image to grayscale
        print('Original image is')
        self.show()
        self.img = np.dot(self.img[...,:3],[0.299,0.587,0.144])
        print("Image after conversion is")
        plt.imshow(self.img, cmap = 'gray')
        plt.show()
      
    
    def bw(self):
        # convert to black n white
        print('Original image is')
        self.show()
        self.img = np.dot(self.img[...,:3],[0.299,0.587,0.144])
        print("Converted image is")
        plt.imshow(self.img, cmap = 'gray')
        plt.show()
    
    def save(self):
        # saving the image
        plt.imsave('new1.jpg', self.img)
        
    def show(self):
        # to display the image
        plt.imshow(self.img)
        plt.show()

    def main(self):
        while True:
            choice = int(input("""Enter your choice:
            1. Load Image
            2. Edit Image
            3. Save Image
            4. Quit"""))
            if choice == 1:
                self.load_image()
                self.show()
                
            elif choice == 2:
                opt = input("""How do you want to edit your image?
                a. Horizontal Flip
                b. Vertical Flip
                c. Invert Colors
                d. Convert to greyscale
                e. Convert to B & W""")


                if opt == 'a':
                    self.horizontal_flip()
#                     self.show()

                elif opt == 'b':
                    self.vertical_flip()
#                     self.show()

                elif opt == 'c':
                    self.invert_color()
#                     self.show()

                elif opt == 'd':
                    self.to_grayscale()

                elif opt == 'e':
                    self.bw()

                else:
                    print("Invalid choice! Try again!")

            elif choice == 3:
                self.save()

            elif choice == 4:
                break
            else:
                print("Invalid choice!!")
                break

image = Image_Manipulation()
image.main()
```
